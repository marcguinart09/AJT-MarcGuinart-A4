```python
#apartat 1
def fibonacci(n):
    succ = [0 for num in range (0,n)]
    succ[0] = 0
    succ[1] = 1
    for n in range(2,n):
        succ[n] = succ[n-1] + succ[n-2]
    return succ

n = 10
fibonacci(n)
```

```python
#apartat 1
def poblacio(p0,r):
    p1=int(p0*(1+r))
    return p1
p0 = int(input("Dona la població inicial"))
r = 0.5
p1 = poblacio(p0,r)
print(f"La població al cap d'un dia serà de {p1} habitants")
```
